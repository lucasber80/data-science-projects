```python
import math
```

```python


def limpar_linha(linha):
  return linha.strip()


def carregar_arquivo(nome_arquivo):
    with open(nome_arquivo, 'r') as f:
        linhas = f.readlines()[1:] 

    linhas = [limpar_linha(linha) for linha in linhas]
    
    return linhas

def retornar_dados_contrato(linha):
  linha_split = linha.split(' ')
  fornecedor = int(linha_split[0])
  mes_inicial = int(linha_split[1])
  mes_final = int(linha_split[2])
  valor = float(linha_split[3])
  return fornecedor,mes_inicial,mes_final,valor

def inicializar_matriz(fornecedores_quant, total_meses):
    matriz_inicial = []
    for _ in range(fornecedores_quant+1):
        fornecedor = []
        for _ in range(total_meses+1):
            mes_inicial = []
            for _ in range(total_meses+1):
                mes_inicial.append(math.inf)
            fornecedor.append(mes_inicial)
        matriz_inicial.append(fornecedor)
    return matriz_inicial


def retornar_total_fornecedores(linhas):
  fornecedores = [linha.split()[0] for linha in linhas]
  fornecedor_quant = len(set(fornecedores))

  return fornecedor_quant

def retornar_total_meses(linhas):
  meses = [int(linha.split()[2]) for linha in linhas]
  meses.sort()
  meses_quant = meses[-1]

  return meses_quant


def retornar_matriz_de_contratos(linhas_de_texto):
  fornecedores_quant = retornar_total_fornecedores(linhas_de_texto)
  total_meses = retornar_total_meses(linhas_de_texto)
  matriz_de_contratos = inicializar_matriz(fornecedores_quant,total_meses)
  for linha in linhas_de_texto:
    fornecedor,mes_inicial,mes_final,valor = retornar_dados_contrato(linha)
    matriz_de_contratos[fornecedor][mes_inicial][mes_final] = valor
  return matriz_de_contratos
```

```python
linhas_de_texto = carregar_arquivo('entradaGrande.txt')
matriz = retornar_matriz_de_contratos(linhas_de_texto)
```

```python
def retornar_maior_intervalo(mes_inicial,mes_final,matriz_contratos):
    maior_intervalo = []
    maior_diferenca = -1
    for idx_fornecedor,fornecedor in enumerate(matriz_contratos):
        for idx_mes_inicial, mes_inicial_lista in enumerate (fornecedor):
            for idx_mes_final,valor in enumerate(mes_inicial_lista):
                diferenca = idx_mes_final - idx_mes_inicial
                if(diferenca > maior_diferenca and not math.isinf(valor) and idx_mes_inicial >= mes_inicial and idx_mes_final <= mes_final):
                    maior_diferenca = diferenca
                    maior_intervalo = [idx_fornecedor,idx_mes_inicial,idx_mes_final,valor]
    return maior_intervalo


def retornar_sequencia_matriz(mes_inicial,mes_final,matriz_contratos):
    contratos = []
    while mes_inicial < mes_final:
        contrato_com_maior_intervalo = retornar_maior_intervalo(mes_inicial,mes_final,matriz_contratos)
        if contrato_com_maior_intervalo == []:
            print('NÃ£o foram encontrados contratos que cumpram essa sequencia de meses')
            break
        contratos.append(contrato_com_maior_intervalo)
        mes_inicial = contrato_com_maior_intervalo[2]+1
    return contratos

retornar_sequencia_matriz(0,4,matriz)
```
